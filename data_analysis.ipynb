## Dataset Analysis

The dataset consists of data corresponding to crimes recorded from the year 2001-2015. The dataset is distributed district-wise for each state of India at the time of collection of data. The original dataset has been appended with the last 3 columns which contain the literacy rate present in the state based on 2001 and 2011 census.

```python
df.shape
```

There are a total of 11,527 entries corresponding to the 15 years of data. As seen above, there are 18 columns (or attributes) present in the data which represents data corresponding to various crimes commited like murder, rape, foeticide etc. Most of the attributes are numerical in nature.

```python
df['STATE'].unique()
```

There are multiple names corresponding to a single state. Also, some names are capitalized. These are preprocessed as follows:

```python
for i in range (len(df)):
    df.loc[i, 'STATE'] = df.loc[i, 'STATE'].lower()
    df.loc[i, 'DISTRICT'] = df.loc[i, 'DISTRICT'].lower()
    if df.loc[i, 'STATE'] == 'a & n islands' : df.loc[i, 'STATE'] = 'a&n islands'
    if df.loc[i, 'STATE'] == 'delhi ut' : df.loc[i, 'STATE'] = 'delhi'
    if df.loc[i, 'STATE'] == 'd & n haveli' : df.loc[i, 'STATE'] = 'd&n haveli'
    if df.loc[i, 'DISTRICT'] == 'total district(s)' : df.loc[i, 'DISTRICT'] = 'total'
    if df.loc[i, 'DISTRICT'] == 'zdelhi ut total' : df.loc[i, 'DISTRICT'] = 'total'
        
df['STATE'].unique()
```

```python
df[['Infanticide', 'Literacy Rate- Total', 'Literacy Rate - Rural', 'Literacy Rate - Urban']].describe()
```

## Missing Values
The columns Literacy Rate - Total, Literacy Rate - Urban and Literacy Race - Rural have only 55 entries. Thus, their data is very sparse and we opt to drop these 3 columns. We can use their values for useful insights during visualization plots later.

Also, the column Total is the sum of values of different crimes in each column and thus is redundant. We are also going to drop it.

The column Infanticide contains data only from the year 2013-2015 and has NA for years 2001-2012. Thus we are dropping it too. We may use it for year specific analysis in years from 2013-2015.

```python
df = df.drop(['Infanticide', 'Literacy Rate- Total', 'Literacy Rate - Rural', 'Literacy Rate - Urban'], axis=1)
```

```python
df.columns
```

Upon closer observation on the dataset, we see that for each state there is a row corresponding to the total amount of crimes in each column. We are also going to remove this entry for each state and for each year. Also the rows containing atleast one NA values are removed.

```python
df = df[df['DISTRICT'] != 'total']
df.dropna(inplace=True)
df.reset_index(drop=True, inplace=True)
```

```python
df.shape
```

## Aggregation
We have thus a total of 10995 entries. However, the dataset is quite sparse with large amount of zeroes. Thus, we perform aggregation and combine multiple entries within each state into a single entry so as to reduce the sparsity of the data. 

However, this combination is not possible on a district wise manner since for combining districts we need to find its neighbouring districts which is not possible based on the given data. Thus, we analyze data and produce results at a state level, disregarding the district in which the crime had occurred.

```python
data_frame = []
i = 0
while i < len(df):
    state = df.loc[i, 'STATE']
    year = df.loc[i, 'Year']
    count = 0
    for j in range(i, i+4):
        if j < len(df) and df.loc[j, 'STATE'] == state and df.loc[j, 'Year'] == year:
            count = count + 1
    temp = df.iloc[i: i+count].drop(['STATE', 'DISTRICT', 'Year'], axis=1).sum().to_list()
    temp.insert(0, state)
    temp.insert(1, int(year))
    data_frame.append(temp)
    if count == 0:
        i = i + 1
    else:
        i = i + count
    
df_columns = list(df.columns[3:14])
df_columns.insert(0, 'State')
df_columns.insert(1, 'Year')
df_aggregated = pd.DataFrame(data_frame, columns=df_columns)
df_aggregated.head(n=50)
```

```python
df_aggregated.shape
```

Thus we see after aggregation, the total number of entries has been reduced from 10995 to 2971. The overall statistics corresponding to these are as follows:

```python
df_aggregated.iloc[:,2:].describe()
```

## Binarization
In order to apply data mining algorithms like associate rule mining, we need to convert our continuous valued data into binary data through the process of binarization.

The threshold value for an attribute j in an entry i is given by:
Let total crimes in entry i for all attributes be p
Let total crimes of all entries for all attributes in the state (for particular year) be z
Let total crimes of all entries for attribute j be w
Then threshold = (p/z)*w

```python
i = 0
while i < len(df_aggregated):
    state = df_aggregated.loc[i, 'State']
    year = df_aggregated.loc[i, 'Year']
    j = i
    count = 0
    
    while j < len(df_aggregated) and df_aggregated.loc[j, 'State'] == state and df_aggregated.loc[j, 'Year'] == year:
        count = count + 1
        j = j + 1
    
    temp = df_aggregated.iloc[i:i+count].drop(['State', 'Year'], axis=1).sum().to_list()
    j = i
    while j < len(df_aggregated) and df_aggregated.loc[j, 'State'] == state and df_aggregated.loc[j, 'Year'] == year:
        for k in range (df_aggregated.shape[1] - 3):
            if df_aggregated.iloc[j][k+2] > ((df_aggregated.iloc[j][12] / temp[10]) * temp[k]):
                df_aggregated.iat[j, k+2] = 1
            else:
                df_aggregated.iat[j, k+2] = 0
        j = j + 1
    i = i + count
    
df_aggregated = df_aggregated.drop(['Total'], axis=1)
```

Thus, now each entry can be considered as a transaction and can be used for association analysis. We further remove all transactions that contain no items, that is all of their entries are zero.

```python
df_binarized = df_aggregated.drop(['State', 'Year'], axis=1)
for i in range (len(df_binarized)):
    var = False
    for j in range(len(list(df_binarized.iloc[i]))):
        if df_binarized.iloc[i][j] != 0:
            var = True
    if var == False:
        df_aggregated = df_aggregated.drop(df.index[[int(i)]])
```
